3) Find top 10 cities with the most amount of funding received.

```python
#Checking for outliers in the Funding Amount column by plotting a boxplot
sns.boxplot(x=df['Funding Amount'])
plt.show()
```

```python
#Creating a function to remove outlier
def outlier(column):
    percentile25=df[column].quantile(0.25)
    percentile75=df[column].quantile(0.75)
    iqr=percentile75-percentile25
    upper_range=percentile75+1.5*iqr
    lower_range=percentile25-1.5*iqr
    return lower_range,upper_range
```

```python
outlier('Funding Amount')
```

```python
#Now removing outlier and creating a new dataframe
df=df[(df['Funding Amount']>-6000000.0) & (df['Funding Amount']<10000000.0)]
df.shape
```

```python
df1=df.groupby('City')['Funding Amount'].sum().reset_index()
df1=df1.sort_values('Funding Amount',ascending=False)[:10]
df1
```

```python
# Top 10 City with most amount of fundings
df1['Percentage']=round((df1['Funding Amount']/df1['Funding Amount'].sum())*100,2)
df1=df1.sort_values('Percentage',ascending=False)
df1
```

```python
#Visualizing using bar chart
fig=plt.figure(figsize=(10,6))
sns.barplot(df1,x='City',y='Percentage')
plt.title("Top 10 Cites With Most Amount Of Fundings Received",fontsize=14,fontstyle='italic')
plt.xticks(rotation=30)
plt.show()
```

```python
#Observations
```
